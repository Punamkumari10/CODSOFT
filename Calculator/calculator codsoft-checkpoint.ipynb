```python
import tkinter as tk
import math

def evaluate_expression():
    try:
        expression = entry.get()
        result = str(eval(expression))
        result_label.config(text="Result: " + result)
    except Exception as e:
        result_label.config(text="Error")

def clear_entry():
    entry.delete(0, tk.END)
    result_label.config(text="Result: ")

def calculate_trig(trig_func):
    try:
        angle = float(entry.get())
        if trig_func == "sin":
            result = math.sin(math.radians(angle))
        elif trig_func == "cos":
            result = math.cos(math.radians(angle))
        elif trig_func == "tan":
            result = math.tan(math.radians(angle))
        result_label.config(text=f"{trig_func}({angle}) = {result:.4f}")
    except ValueError:
        result_label.config(text="Invalid input")
        
app = tk.Tk()
app.title("Calculator")

entry = tk.Entry(app, width=40, borderwidth=5)
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

button_values = [
    ("7", 1, 0), ("8", 1, 1), ("9", 1, 2), ("/", 1, 3),
    ("4", 2, 0), ("5", 2, 1), ("6", 2, 2), ("*", 2, 3),
    ("1", 3, 0), ("2", 3, 1), ("3", 3, 2), ("-", 3, 3),
    ("0", 4, 0), (".", 4, 1), ("=", 4, 2), ("+", 4, 3),
]

for (val, row, col) in button_values:
    button = tk.Button(app, text=val, padx=20, pady=20, command=lambda v=val: entry.insert(tk.END, v))
    button.grid(row=row, column=col)

enter_button = tk.Button(app, text="Enter", bg="green", padx=20, pady=20, command=evaluate_expression)
enter_button.grid(row=5, column=0)

clear_button = tk.Button(app, text="Clear", bg="red", padx=20, pady=20, command=clear_entry)
clear_button.grid(row=5, column=1)

sin_button = tk.Button(app, text="sin", padx=20, pady=20, command=lambda: calculate_trig("sin"))
sin_button.grid(row=5, column=2)

cos_button = tk.Button(app, text="cos", padx=20, pady=20, command=lambda: calculate_trig("cos"))
cos_button.grid(row=5, column=3)

tan_button = tk.Button(app, text="tan", padx=20, pady=20, command=lambda: calculate_trig("tan"))
tan_button.grid(row=6, column=2)

result_label = tk.Label(app, text="Result: ", padx=20, pady=20)
result_label.grid(row=6, column=0, columnspan=2)

app.mainloop()
```
